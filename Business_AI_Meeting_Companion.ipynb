```python
import requests

# Sample audio file download
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMSkillsNetwork-GPXX04C6EN/Testing%20speech%20to%20text.mp3"
r = requests.get(url)
with open("sample.mp3", "wb") as f:
    f.write(r.content)

print("Audio downloaded successfully 🎵")
```

```python
import torch
from transformers import pipeline

pipe = pipeline("automatic-speech-recognition", model="openai/whisper-tiny.en", chunk_length_s=30)
result = pipe("sample.mp3", batch_size=8)["text"]
print("🎙️ Transcribed Text:\n", result)
```

```python
import torch
from transformers import pipeline
import gradio as gr

def transcript_audio(audio_file):
    pipe = pipeline("automatic-speech-recognition", model="openai/whisper-tiny.en", chunk_length_s=30)
    result = pipe(audio_file)["text"]
    return result

iface = gr.Interface(fn=transcript_audio,
                     inputs=gr.Audio(sources="upload", type="filepath"),
                     outputs="text",
                     title="Audio Transcription App",
                     description="Upload any audio and get text.")
iface.launch()
```

```python
from transformers import pipeline
import gradio as gr

# Summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")

def meeting_summary(audio_file):
    # Step 1: Transcribe
    stt = pipeline("automatic-speech-recognition", model="openai/whisper-tiny.en")
    text = stt(audio_file)["text"]

    # Step 2: Summarize
    summary = summarizer(text, max_length=120, min_length=30, do_sample=False)[0]["summary_text"]
    return f"🗣️ Transcription:\n{text}\n\n📋 Summary:\n{summary}"

iface = gr.Interface(fn=meeting_summary,
                     inputs=gr.Audio(sources="upload", type="filepath"),
                     outputs="text",
                     title="Business AI Meeting Companion",
                     description="Upload meeting audio → Get transcription + summary")
iface.launch()
```
